```
#importing Pandas
import pandas as pd
```

```
#reading the CSV file.
df = pd.read_csv('/content/sample_data/file.csv')
```

```
#checking for duplicate records(If there is no duplicate records then all of the values are displayed as False.)
df.duplicated()
```

```
#For checking duplicates for particular coloumn
df.duplicated('longitude')
```

```
#for dropping duplicates
df.drop_duplicates()
```

```
#for dropping duplicates from particular column
df.drop_duplicates(['longitude'])
```

```
#to keep the first record
df.drop_duplicates(['longitude'], keep='first')
```

```
#to keep the last record
df.drop_duplicates(['longitude'], keep='last')
```

```
#to drop all duplicates altogether
df.drop_duplicates(['longitude'], keep=False)
```
